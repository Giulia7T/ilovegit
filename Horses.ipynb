```python
import pandas
import matplotlib.pyplot as plt
from scipy.stats import poisson
import numpy as np
```

```python
Prussians = pandas.DataFrame()
Prussians["deaths"] = [0, 1, 2, 3, 4]
Prussians["groups"] = [109, 65, 22, 3, 1]
Prussians
```

```python
N_men = 122
N_groups_peryear = 10
N_years = 20
N_groups = N_groups_peryear*N_years
```

```python
plt.bar(Prussians["deaths"], Prussians["groups"]/N_groups, color="black")
```

```python
mu = sum([Prussians["deaths"][i]*Prussians["groups"][i] for i in range(len(Prussians["deaths"]))])/N_groups
sigma = sum([((Prussians["deaths"][i]-mu)**2)*Prussians["groups"][i] for i in range(len(Prussians["deaths"]))])/N_groups
```

```python
plt.scatter(Prussians["deaths"], Prussians["groups"]/N_groups, color="black")
x = np.linspace(0, 4, 10)
plt.plot(Prussians["deaths"], poisson(mu).pmf(Prussians["deaths"]), color="coral")
```
