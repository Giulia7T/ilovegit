```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import norm
```

```python
N = 5
mu = 1
sigma = 0.2
positions = np.random.normal(loc=mu, scale=sigma, size=N)
plt.hist(positions, color="orangered")
plt.xlabel("Quasars positions")
```

```python
x = np.linspace(min(positions), max(positions), 100)
#x = np.linspace(0, 2, 100)
pdfs = []
L = 1
for i in range(len(positions)):
    pdfs.append(norm(positions[i], sigma).pdf(x))
    plt.plot(x, pdfs[i])
    L = L*norm(positions[i], sigma).pdf(x)
```

```python
for i in range(len(positions)):
    plt.plot(x, pdfs[i])

plt.plot(x, L, color="red")
plt.title("likelihood")
plt.xlabel("x")
plt.ylabel("L")
```

```python
plt.plot(x, np.log(L), color="red")
plt.title("log likelihood")
plt.xlabel("x")
plt.ylabel("ln(L)")
```

```python
sorted_indices = np.argsort(L)
index_max = sorted_indices[-1]
x[index_max]
```
