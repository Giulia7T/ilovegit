```python
###MONTY HALL PROBLEM
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
doors = np.array([0, 1, 0], dtype="bool")
```

```python
N = 100 #at higher orders it is always the same
conservative = 0
switcher = 0
last = 0
for i in range(N):
    choice = random.randint(0, 2)
    if doors[choice]:
        conservative += 1
        
    #remaining doors
    rem = np.delete(doors, choice)
    #goat door: np.where(rem == False)[0][0] ...and the switcher chooses the other one
    if np.delete(rem, np.where(rem == False)[0][0])[0]:
        switcher += 1
    
    #doors for the last one are just one car and one goat
    if random.randint(0, 1):
        last += 1
```

```python
people = ["conservative", "switcher", "newcomer"]
plt.bar(people, [conservative/N, switcher/N, last/N], color="purple")
plt.title("Monty Hall simulation")
```

```python
###LOG PDF
```

```python
n = 1000
x = np.random.uniform(0.1, 10, n)
```

```python
plt.hist(x, int(n/50), color="coral")
plt.xlabel("x")
```

```python
y = np.log10(x)

plt.hist(y, int(n/50), color="lightgreen")
plt.xlabel("y")
```

```python
#p(y)=(dx/dy)p(x)=((dy/dx)^-1)p(x)
#p(x)=1/(10-0.1)
#x=10^y
points = np.linspace(-1, 1, 100)
plt.hist(y, int(n/50), color="lightgreen", density=True)
pdf = (10**points)/np.log(10)
pdf_n = pdf/(10-0.1)
plt.plot(points, pdf, color="orangered")
plt.xlabel("y")
```

```python
print("log(mean(x)) = ", np.log10(np.mean(x)))
print("mean(log(x)) = ", np.mean(y))
```

```python
print("log(median(x)) = ", np.log10(np.median(x)))
print("median(log(x)) = ", np.median(y))
```
