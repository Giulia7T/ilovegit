```python
import random
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#cloudy = 0
#clear = 1
clear_given_cloudy = 0.5
cloudy_given_cloudy = 0.5
cloudy_given_clear = 0.1
clear_given_clear = 0.9

today = 0
N_days = 10000
tomorrow = np.zeros(N_days)
```

```python
for i in range(N_days):
    p = random.uniform(0, 1)
    if today == 0:
        if p <= 0.5:
            tomorrow[i] = 0
        if p > 0.5:
            tomorrow[i] = 1
    if today == 1:
        if p <= 0.1:
            tomorrow[i] = 0
        if p > 0.1:
            tomorrow[i] = 1
    today = tomorrow[i]
```

```python
#clear days over the whole period
np.sum(tomorrow)/N_days
```

```python
#cloudy days over the whole period
1-np.sum(tomorrow)/N_days
```

```python
days = np.arange(1, N_days+1, 1)
clear_days = np.zeros(N_days)
for i in days:
    clear_days[i-1] = np.sum(tomorrow[:i])/i
```

```python
plt.plot(days, clear_days, color="coral")
plt.xlabel("day")
plt.ylabel("clear fraction")
```

```python
plt.hist(clear_days[1000:], bins=100, color="coral", histtype="step")
plt.xlabel("clear fraction")
```
