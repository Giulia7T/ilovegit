```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
Prussians_keys = ["deaths", "groups"]
Prussians_values = np.array(([0, 1, 2, 3, 4], [109, 65, 22, 3, 1]))
Prussians = {k:v for k, v in zip(Prussians_keys, Prussians_values)}

for key in Prussians_keys:
    print(key, end="\t")
print()
for i in range(len(Prussians_values[0])):
    print(Prussians_values[0, i], "\t", Prussians_values[1, i])
```

```python
N_men = 122 #total men
N_groups_peryear = 10
N_years = 20
N_groups = N_groups_peryear*N_years #total groups to normalise
```

```python
plt.bar(Prussians["deaths"], Prussians["groups"]/N_groups, color="black")
plt.xlabel("deaths per group")
plt.ylabel("fraction of groups")
plt.title("probability distribution")
```

```python
#estimating the mean...
mu = np.sum(Prussians["deaths"]*Prussians["groups"])/N_groups
```

```python
#trying to fit
plt.plot(Prussians["deaths"], Prussians["groups"]/N_groups, "ok", label="data")
plt.plot(Prussians["deaths"], poisson(mu).pmf(Prussians["deaths"]), "--g", label="fit")
plt.xlabel("deaths per group")
plt.ylabel("fraction of groups")
plt.title("probability distribution")
plt.legend()
```
